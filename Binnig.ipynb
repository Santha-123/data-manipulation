QUANTILE BINNING

```python
df['weight_bin']=pd.qcut(df['Weight'],q=3,labels=['Low','Medium','High'])
df.head(60)
#note:Equal depth binning is also known as quantile binning.
#note contd:It divides a numerical variable into bins of approximately equal size based on the variables distribution.
#note contd:This techniques assign each value to a bin such that each bin contains an equal number of observations,ensuring balanced representation across bins and  aiding in statistical analysis and visualization of data distribution.
```

```python
#equal depth binning
df ['weight_bin'] = pd.qcut(df['Weight'],q=3,labels=['low','medium','high'])
df.head()
#note contd:Equal depth binning is also called as quantile binning,divides numerical variable into bins of approximately equal size based on the variable's distribution.
#note contd...:This techniques assignes each value to a bin contains equal number of observations,ensuring balanced representation across bins and aiding in statistical analysis and visualization of data distribution.
```

```python
#custom binning
df ['weight_bin'] = pd.cut(df['Weight'],bins=[0,1000,2000,3000,4000,5000],labels=['S','M','L','XL','XXL'])
df.head()
#NOTE:Panda'cut' function is used for binning numerical data into discrete intervals or bins.It allows defining bin edges or the number of bins categorizing data baed on specific criteria like range or precision.
#note contd...:'cut' facilitates data segmentation,useful for grouping analyzing continous variables by converting them into categorical data for easier implementation and statistical analysis.
```
