DATAFRAME

```python
Apple=['apple',45]
orange=['orange',34]
pineapple=['pineapple',56]


df=pd.DataFrame([Apple,orange,pineapple])
df
#note:while we declaring datframe we must declare D and F in caps 
#output:I called all the list in one variable name for viewing data easily
#NOTE:Most probably use dataframe df as variable name to view where all the software project developer use this in pandas and you can use df in different names like df_panda,df_fruits
```

```python
#set column names
df.columns = ['fruits','price']
df
#output:defining colunmn name for the above series
```

```python
#write a datframe to file
job1 =['design engineer',4,'lmw','chennai']
job2 =['lead engineer',3,'schennider electric','bangalore']
job3 =['assistant manager',5,'lena meena agencies','cuddalore']
#note:always while defining varaible name you should not leave space
experience=[job1,job2,job3]
df_exp=pd.DataFrame(experience)
df_exp
```

```python
#adding coumns for the above experience dat
df_exp.columns=('position','no.of years experienced','company','location')
df_exp
#output:while adding columns you must assign '=' operator only it also accepts tuple,list,dictionary
```

```python
df_exp.to_csv('experience.csv',index=False)
df_exp
#output:It create the new file automatically by code and it also ignores index when we give index=false
#note:to_csv is the code to create a new file in excel microsoft
#note:index=False i.e F must be in caps to hide index and we can view in the excel sheet new file
```

```python
#read a csv file
df=pd.read_csv('cars.csv')
df
#output:We can read by using this syntax and it will not print all rows it dispalys first some rows and last some rows with total columns and rows
#output: this command is to raed only csv file with extension:csv
```

```python
#read a excel file
df=pd.read_excel('cars.xlsx')
df
#output:We can read by using this syntax and it will not print all rows it dispalys first some rows and last some rows with total columns and rows
#output: this command is to raed only excel file with extension:xlsx
```

```python
#adjusting the header
df=pd.read_excel('cars.xlsx',header=1)
df
#output:By using this header we can hide the first row the unnamed rows in before excel file
```

```python
#read the second sheet in excel
df=pd.read_excel('cars.xlsx',sheet_name='Summary',header=1)
df
#OUTPUT:In Summary S is in caps in the downloaded excel sheet so here too we must give in caps only and we hided the header of the firstrow so we did not get unnamed names at all
```
