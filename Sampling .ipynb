```python
#import library and load dataset
#import library
import pandas as pd
```

```python
#load dataset
df=pd.read_csv('cars.csv')
df
```

```python
#undersampling
df.sample (n= 100)
#note:Undersampling in pandas involves reducing the size of the dataframe by randomly selecting a subset of rows,typically from the majority class,to a balance class distributions.
#note cont..:This techniques helps mitigate class imbalance within datsets improved model training and evaluation,addressing challenges such as bias towards the dominent class in machine learning tasks.
```

```python
#sampling by fraction
df.sample(frac=0.3)
```

```python
#oversampling
df.sample(n=1000,replace=True)
#NOTE:In true 'T' must be in caps if not it wil not be defined and occurs error.
#note:Oversampling in pandas involves increasing a size of a dataframe by duplicating or generating new rows,usually from minority class to balance class distributions.
#note contd..:This techniques helps alleviate class imbalance within datsets ,enhancing the representation of underrepresented classes for more robust model training and accurate predictions in machine learning tasks.
```
