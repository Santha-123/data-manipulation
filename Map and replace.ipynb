```python
#import library and load dataset
#import library
import pandas as pd
```

```python
#load dataset
df=pd.read_csv('cars.csv')
df
```

Map values

```python
#map function
df['Origin'] = df['Origin'].map({
    'US':1,
    'Europe':2,
    'Japan':3
})
df.head()
#OUTPUT:Instead of US,Europe and Japan it replaces according to corresponding number where we assigned.
#note:In pandas the map() function is used to transform values in a series or dataframe column based on a mapping relationship defined by a dictionary,a series or a function 
#not contd...:It provides a connvenient ay to apply custom-transformation element_wise ,replacing values according to specified mappings or functions,enhancing data clarity and analysis flexibility
```

Replace a  value

```python
#replace
df=pd.read_csv('cars.csv')
df.head(2)
```

```python
df['Origin'] = df['Origin'].replace('US','USA')
df.head()
#output:It replaces 'USA' intead of 'US'.
#NOTE:In pandas the replace() function is used to replace values in a series or dataframe with other values.
#note contd...:It accepts dictionaries,list or single values as replacements,allowing for versatile data manipulation by substiting specific values with new ones across specified columns or the entire dataset,facilitating data cleaning and standardization tasks efficiently   
```
