Import library and load datset

```python
#import library
import pandas as pd
```

```python
#load dataset
df=pd.read_csv('cars.csv')
df
```

```python
#applying condition on series
df['Weight']>=3500
#note:I asked to print weight which is greater than 3500 and it prints true or false only when we use location only it displays from that table
```

```python
#Filtering rows based on conditions
df.loc[df['Weight']>=3500]
#note:we used df.loc to display weight condition from the table
```

```python
#filtering rows using string value
df.loc[df['Origin']== 'US']
#note:I dislplayed the origin which has only US and I used loc funcion so it displays from the table
```

```python
#filtering rows using integer value
df.loc[df['Model']== 70]
#note:I dislplayed the model which has only 70 and I used loc funcion so it displays from the table
```
