```python
#import library and load dataset
#import library
import pandas as pd
```

```python
#load dataset
df=pd.read_csv('cars.csv')
df
```

```python
#pivot-rows to columns
df=df.drop_duplicates('Car')
df.pivot(index='Car',columns='Origin',values='Weight')
#NOTE:PIVOT is a data restructuring operation that transforms data from long to wide format,organizing along with two or more axes.It involves selecting an index(rows),columns and values to reshape data for clear analysis and presentation.It is useful in summarizing and aggregating information for better insight
```

```python
#pivot table
#I first read the table
df=pd.read_csv('cars.csv')
df
```

```python
df.pivot_table(values='Weight',index='Origin',columns='Model',aggfunc='max')
#note:pivot table is a multidimensional data that summarizes data from a large dataset.It allows users to dynamically arrange and aggregate data by choosing rows,columns,and values to analyze,providing insights to computations like sums,averages or counts.Pivot table are widely used in data analysis to condense and visualize complex information efficiently.
```

```python
#melt-column to rows
pd.melt(df,id_vars=['Car'],value_vars=['Origin','Weight'])
#note:melt is a data transformation method in pandas that unpivots a dataframe from wide to long format,reshaping it by converting column into rows.
#note contd...:It gathers multiple columns into two key columns:one containing variable names another with corresponding values,aiding in analysis and visualization of data relationships across different variables
```
